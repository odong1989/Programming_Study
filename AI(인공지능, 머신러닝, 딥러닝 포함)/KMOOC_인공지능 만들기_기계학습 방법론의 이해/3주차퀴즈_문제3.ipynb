```
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

#데이터 불러오기
x = ['male','male','male','male','male','female','female','female','female','female']
y = [156, 374, 457, 567, 364, 256, 369, 478, 467, 500]

#dummy(1,-1) 코딩
x_dummy = list(map(lambda x : 1 if x == 'female' else -1, x))
df = pd.DataFrame( {'y':y, 'x_dummy':x_dummy}, columns = ['y','x_dummy'])
df
```
