```
from sklearn.datasets import load_boston
```

```
from sklearn.linear_model import LinearRegression, Ridge, Lasso
# 선형회귀 : LinearRegression
# 리지회귀 : Ridge
# 라쏘회귀 : Lasso
```

```
boston = load_boston()
X = boston.data
y = boston.target
```

```
boston
```

```
X #X = boston.data
```

```
y #y = boston.target
```

```
#1.선형회귀(LinearRegression)
```

```
#학습을 하는 인스턴스 lr1를 생성.
lr1 = LinearRegression()
lr1
```

```
#lr1 인스턴스에 대해 학습을 하는 fit메소드를 X(특징 벡터의 집합)과 y(정답 정보)를 인수로 불러낸다.
lr1.fit(X,y)
#이를 통해 선형회귀식이 얻어진다.
#예측하고 싶은 데이터(13차원 벡터 x)를 인수로 하여 predict 메소드를 불러내면 예측값을 출력해준다.
```

```
#정칙화의 효과를 확인해본다.
#학습한 선형회귀식의 계수와 그 제곱합을 표시하는 코드이다.
print("Linear Regression")
for f, w in zip(boston.feature_names, lr1.coef_) :
    print("{0:7s}: {1:6.2f}".format(f,w))
print("coef = {0:4.2f}".format(sum(lr1.coef_**2)))
```

```
#2.리지회귀(Ridge)
```

```
lr2 = Ridge(alpha=10.0) #책에서는 정칙화항의 가중치를 10.0으로 한다고만 설명.
lr2.fit(X,y)

print("Ridge")
for f, w in zip(boston.feature_names, lr2.coef_) :
    print("{0:7s}: {1:6.2f}".format(f,w))
print("coef = {0:4.2f}".format(sum(lr2.coef_**2)))
```

```
#3.라쏘회귀
```

```
#alpha(정칙화항의 가중치)를 2.0으로하면 확실히 몇개의 계수가 0이 되어 있는 것을 알 수 있다.
lr3 = Lasso(alpha=2.0)
lr3.fit(X,y)
print("Lasso")
for f, w in zip(boston.feature_names, lr3.coef_) :
    print("{0:7s}: {1:6.2f}".format(f,w))
print("coef = {0:4.2f}".format(sum(lr3.coef_**2)))
```
