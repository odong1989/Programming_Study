```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 100) #100등은 픽셀 수라고 생각할수 있다. 많으면 많을수록 곡선이 부드럽다.
C = np.cos(X)

plt.title("x,y 축 범위 설정")
plt.plot(X,C)
plt.xticks([-np.pi, np.pi /2 ,0, np.pi/2 , np.pi]) #원하는 눈금을 정확히 보여주게 하려면 이 설정을 해야한다.
plt.yticks([-1,0,1])                               # 원하는 눈금을 정확히 보여주게 하려면 이 설정을 해야한다.
plt.show()
```

```python
#범례 
X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X) , np.sin(X)

plt.title("legend를 표시한 플롯")
plt.plot(X,C,ls="--",label="cosine")
plt.plot(X,S,ls=":",label="sine")
plt.legend(loc=0)
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X) , np.sin(X)

plt.plot(X,C,label="cosine")
plt.xlabel("time")
plt.ylabel("sipal")
plt.plot(X,S,ls=":",label="sine")
plt.legend(loc=0)
plt.show()
```

```python
np.random.seed(2)
f1 = plt.figure(figsize=(10,2))
plt.plot(np.random.randn(100))
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1=np.cos(2*np.pi*x1) *np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.title('time (s)')
plt.ylabel('undamped')
print(ax2)

plt.tight_layout()
plt.show()
```
