```
#파이썬의 값 할당방법은 8가지로 나뉨.
#①기본할당
#②같은 값 할당
#③각각 해당 값 할당
#④값 바꾸기
#⑤값 증가
#⑥값 감소
#⑦정의는 되었지만 값 없음
#⑧변수 제거
```

```
#①기본할당
x=1+1.2
x
```

```
#②같은 값 할당
x= y= 0
[x,y]
```

```
#③각각 해당 값 할당
x,y = 1, 0.9
[x,y]
```

```
#④값 바꾸기
x,y = y,x
[x,y]
```

```
#⑤값 증가
x = 0
x +=2
x
```

```
#⑥값 감소
x= 0
x -= 3
x
```

```
#⑦정의는 되었지만 값 없음
x= None
[x, type(x)]
```

```
#⑧변수 제거
del(x)
x
```
