```
#[1]기본데이터형
#①int
#②float
#③bool
#④str
#⑤bytes

#[2]컨테이너형(컨테이너형:기본 데이터형을 하나의 요소로 하여 만들어지는 데이터 형)
#①list
#②tuple
#③dict
#④set
#⑤str
#⑥bytes
#*  str,bytes 데이터 형은 컨테이너 데이터형도 될 수 있다.
#** tuple, str, bytes는 값을 변경할 수 없다.
```

```
#[1]기본데이터형
#①int
#②float
#③bool
#④str
#⑤bytes
```

```
#int
x = 100
type(x)
```

```
#float
x = 0.0
type(x)
```

```
#bool
x=True
type(x)
```

```
#str
x= 'hello! \n Welcome to ML world'
type(x)
```

```
# bytes
x= b'bong'
type(x)
```

```
#[2]컨테이너형(컨테이너형:기본 데이터형을 하나의 요소로 하여 만들어지는 데이터 형)
#①list
#②tuple
#③dict
#④set
```

```
#list
x = [1,2,3]
type(x)
```

```
#리스트와 각 원소의 데이터 형
x = ["a",2.3, 3, b"bong"]
print(type(x))
print(type(x[0]))
print(type(x[1]))
print(type(x[2]))
print(type(x[3]))
```

```
#tuple
x= (1,2,3)
print(type(x))
```

```
#dict
x = {"key1":100, "key2":200}
type(x)
```

```
x = dict(key1=100, key2=200)
type(x)
```

```
#set
x= {"key1","key2"}
type(x)
```

```
x= {1,2,3,3}
x
```
