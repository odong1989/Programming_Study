```
#문자열에 대한 연산방법 종류
#①공백 제거
#②지정 문자의 출현횟수 구하기
#③조회
#④대소문자 변환
#⑤분리 및 결합
#⑥


#①strip
#②count
#③index
#④is
#⑤upper
#⑥lower
#⑦split
#⑧join
```

```
#문자열 생성
name = '\n bong ju kang'
name
```

```
#strip() : 문자열 앞뒤 화이트 스페이스 삭제
name.strip()
```

```
#문자열의 시작하는 인덱스 : index
name.index('k')
```

```
#대소문자 변환 : lower upper
['KANG'.lower(), 'banene'.upper()]
```

```
#대소문자 서로 변환 : swapcase
'Kang, Bong Ju'.swapcase()
```

```
#숫자, 문장로만 구성된 문자열인지 확안 : islanum
'Bonjuor Kang 2020!'.isalnum()
```

```
#문자열 분리 : split
'Kang, Bonjour'.split(',')
```

```
#문자열 결합
','.join(['Kang', 'Bong Ju'])
```
