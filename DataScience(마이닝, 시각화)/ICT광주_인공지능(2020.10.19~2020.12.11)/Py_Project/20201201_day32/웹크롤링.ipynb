```python
#웹크롤링 예제3-네이버의 영화랭크 가져와보기
import requests
response = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn')
html = response.text

print(html)
#네이버 무비 랭킹을 가져와서 텍스트로 화면에 표시
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser') #html.parser를 사용하여 soup에 넣겠따.
#div[class=tit3]인 애들만 선택해서 출력하겠다.
Title_html = soup.select('div[class=tit3]')
print(Title_html)
```

```python
#영화이름만 나오도록 설정
for tag in soup.select('div[class=tit3]'):
    print(tag.text)
```

```python
#영화정보(문자)만 보기
for tag in soup.select('div[class=tit3]'):
    print(tag.text.split())
```

```python
#파일에 저장-나의버전
outFp = open("data20201201_movie.txt","w",encoding='utf-8')

for tag in soup.select('div[class=tit3]'):
    #print(tag.text.split())
    outFp.writelines(tag.text.split())
outFp.close()
print("-----정상적으로 파일이 복사되었음----")
```
